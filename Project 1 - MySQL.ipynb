### Question 2

SELECT Continent, count(*) AS Number_Countries, sum(population) AS Population <br> 
FROM country <br>
WHERE population > 0 <br>
GROUP BY Continent <br>
ORDER BY 1 ASC <br>

```python
query = '''SELECT Continent, count(*) AS Number_Countries, sum(population) AS Population 
            FROM country 
            WHERE population > 0 
            GROUP BY Continent 
            ORDER BY 1 ASC'''
pd.read_sql_query(query, con=cnx, index_col = "Continent")
```

```python
df2 = country[country.Population > 0][['Continent', 'Population']].groupby('Continent').agg(['count', 'sum']).sort_index()
df2.columns = ['Number_Countries', 'Population']
df2
```

### Question 3

SELECT city.Name AS City, city.population <br>
FROM city <br> 
INNER JOIN country ON city.CountryCode = country.code <br>
WHERE country.code = 'USA' <br>
ORDER BY city.population DESC LIMIT 10 <br> 

```python
query = '''SELECT city.Name AS City, city.population
        FROM city 
        INNER JOIN country ON city.CountryCode = country.code 
        WHERE country.code = 'USA' 
        ORDER BY city.population DESC LIMIT 10;'''

pd.read_sql_query(query, con=cnx)
```

```python
df3 = pd.merge(country, city, left_index=True, right_index = True)
df3 = df3[df3.index == 'USA'][['Name_y', 'Population_y']].sort_values(by = "Population_y", ascending = False).head(10)
df3.columns = ['City', 'Population']
df3.reset_index(drop = True)
```

### Question 4

SELECT country.Name, Language, (Percentage * population) / 100 <br> 
FROM countrylanguage <br>
INNER JOIN country ON countrylanguage.CountryCode = country.code <br>
WHERE IsOfficial = True <br>
ORDER BY 3 DESC limit 10 <br>

```python
query = '''SELECT country.Name, Language, (Percentage * population) / 100 
            FROM countrylanguage 
            INNER JOIN country ON countrylanguage.CountryCode = country.code 
            WHERE IsOfficial = True 
            ORDER BY 3 DESC limit 10 '''

pd.read_sql_query(query, con=cnx)
```

```python
df4 = pd.merge(country, countrylanguage, left_index=True, right_index = True)
df4['Percentage * population / 100'] = (df4.Percentage * df4.Population)/100
df4[df4.IsOfficial == 'T'][["Name","Language", "Percentage * population / 100"]]\
    .sort_values(by = "Percentage * population / 100", ascending = False).head(10).reset_index(drop = True)
```

### Question 5

SELECT Language, sum((Percentage * population) / 100) <br> 
FROM countrylanguage <br>
INNER JOIN country ON countrylanguage.CountryCode = country.code <br>
GROUP BY Language <br>
ORDER BY 2 DESC <br>
LIMIT 5; <br>

```python
query = '''SELECT Language, sum((Percentage * population) / 100) 
            FROM countrylanguage 
            INNER JOIN country ON countrylanguage.CountryCode = country.code 
            GROUP BY Language 
            ORDER BY 2 DESC 
            LIMIT 5; '''
pd.read_sql_query(query, con=cnx)
```

```python
df5 = pd.merge(countrylanguage, country, left_index=True, right_index = True)
df5['sum((Percentage * population) / 100)'] = (df5.Percentage * df5.Population)/100
df5[["Language", "sum((Percentage * population) / 100)"]].groupby('Language').sum()\
.sort_values(by = "sum((Percentage * population) / 100)", ascending = False).head(5).reset_index()
```
